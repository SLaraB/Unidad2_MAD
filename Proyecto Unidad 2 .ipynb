```python
territory = gpd.read_file(SHAPEFILE, encoding='utf-8')
territory = territory[territory["codregion"] == REGION]
# Obtenemos los datos de las Columnas "Comuna", "geometry"
territory = territory[["Comuna", "geometry"]]
```

```python
def save_map(df, territory, image_name: str = "Mapa_Calor"):
    
    #Junta las tablas df y Pais
    territory = territory.merge(df, on = "Comuna")
    
    #Inicializa matplotlib para la creacion del mapa
    fig, ax = plt.subplots(figsize=(10, 10))
    max_presente = df.max()[1]#valore maximo para la barra de calor
    territory.plot(column = 'Valor',ax=ax,edgecolor = "gray", cmap = 'OrRd', legend = False).set_axis_off()
    bar_info = plt.cm.ScalarMappable(cmap="Reds", norm=plt.Normalize(vmin=0,vmax=max_presente))
    bar_info._A = []
    cbar = fig.colorbar(bar_info)
    
    #Guardar imagen en nombre_imagen
    plt.savefig('.//{}.svg'.format(image_name), bbox_inches = "tight",transparent = True)
```

```python
lista = []
for i, row in territory.iterrows():
    comuna = row["Comuna"]
    valor = df_por_comuna.loc[i, 'Lago_Ranco':'Valdivia'].sum().astype(int)
    lista.append({"Comuna": comuna, "Valor": valor})

df_comuna_mapa = pd.DataFrame(lista)
save_map(df_comuna_mapa, territory, "mapa")
ruta_mapa = os.path.join(carpeta_graficos, "mapa_calor_noticias.png")
plt.title("Mapa de calor noticias incendios forestales en Los Rios")
plt.savefig(ruta_mapa, dpi=500)
```

```python
def find_keywords(text):
    keyword_locations = {}
    for word in KEYWORDS_ENCENDIO:
        if word in text.lower():
            keyword_locations[word] = [index for index, value in enumerate(text.lower().split()) if value == word]
    return keyword_locations
```

```python
df_sentimiento_respuesta = dataset_filtrado_respuesta

df_sentimiento_respuesta['prediction'] = ""
df_sentimiento_respuesta['score'] = ""
df_sentimiento_respuesta
```

```python
for index, row in tqdm(df_sentimiento_respuesta.iterrows(), desc='Análisis de sentimiento', total=df_sentimiento_respuesta.shape[0]):
    sentiment_value = sentiment_analyzer.predict(row['title'])
    
    # Insertamos en el DataFrame
    df_sentimiento_respuesta.at[index, "prediction"] = sentiment_value.output
    df_sentimiento_respuesta.at[index, "score"] = sentiment_value.probas
```
